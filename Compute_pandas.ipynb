```python
df['artist'].value_counts().head(10) #finding the top 10 most occuring artist
```

```python
top10=df['artist'].value_counts()[:10].index.tolist() #putting the top 10 most occuring artist in a list(top10)
```

```python
df10 = pd.DataFrame(top10, columns=['artist']) #creating a dataframe with top 10 most occuring artist
```

```python
df10['rep']=df['artist'].value_counts()[:10].tolist() #puttung number of times an artist is repeated
```

```python
new_tempo=[] #finding the average tempo of all the artist in the top10 list
for i in range(10):
  t=df[df['artist']==df10.iloc[i,0]]['tempo'].mean()
  new_tempo.append(t)
```
